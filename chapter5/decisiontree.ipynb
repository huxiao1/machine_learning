```python
#load data
import pandas as pd
import numpy as np
data = pd.read_csv('iris_data.csv')
data.head()
```

```python
#define x and y
x = data.drop(['target','label'],axis=1)
y = data.loc[:,'label']
y.head()
print(x.shape,y.shape)
```

```python
#establish the decision tree
from sklearn import tree
dc_tree = tree.DecisionTreeClassifier(criterion='entropy',min_samples_leaf=5)
dc_tree.fit(x,y)
```

```python
#evalue the model
y_predict = dc_tree.predict(x)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y,y_predict)
print(accuracy)
```

```python
#visualize the tree
%matplotlib inline
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(20,20))
tree.plot_tree(dc_tree,filled='True',feature_names=['SepalLength','SepalWidth','PetalLength','PetalWidth'],
               class_names=['setosa','versicolor','virginic'])
```

```python
#establish the decision tree
from sklearn import tree
dc_tree2 = tree.DecisionTreeClassifier(criterion='entropy',min_samples_leaf=1)
dc_tree2.fit(x,y)
#visualize the tree
%matplotlib inline
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(20,20))
tree.plot_tree(dc_tree2,filled='True',feature_names=['SepalLength','SepalWidth','PetalLength','PetalWidth'],
               class_names=['setosa','versicolor','virginic'])
```
