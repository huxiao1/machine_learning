```python
#load the data
import pandas as pd
import numpy as np
data = pd.read_csv('anomaly_data.csv')
data.head()
```

```python
#visualize the tree
%matplotlib inline
from matplotlib import pyplot as plt
plt.scatter(data.loc[:,'x1'],data.loc[:,'x2'])
plt.title('data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
#define x1,x2
x1 = data.loc[:,'x1']
x2 = data.loc[:,'x2']
```

```python
fig2 = plt.figure(figsize=(10,5))
plt.subplot(121)
plt.hist(x1,bins=100)
plt.title('x1 distribution')
plt.xlabel('x1')
plt.ylabel('counts')
plt.subplot(122)
plt.hist(x2,bins=100)
plt.title('x2 distribution')
plt.xlabel('x2')
plt.ylabel('counts')
plt.show()
```

```python
#calculate mean and sigma of x1 and x2
x1_mean = x1.mean()
x1_sigma = x1.std()
x2_mean = x2.mean()
x2_sigma = x2.std()
```

```python
#calculate the gaussian distribution p(x)
from scipy.stats import norm
x1_range = np.linspace(0,20,300)
normal1 = norm.pdf(x1_range,x1_mean,x1_sigma)
plt.plot(x1_range,normal1)
```

```python
x2_range = np.linspace(0,20,300)
normal2 = norm.pdf(x2_range,x2_mean,x2_sigma)
plt.plot(x2_range,normal2)
```

```python
#establish the model and predict
from sklearn.covariance import EllipticEnvelope
clf = EllipticEnvelope()
clf.fit(data)
y_predict = clf.predict(data)
```

```python
fig = plt.figure(figsize=(10,5))
mormaly_points=plt.scatter(data.loc[:,'x1'],data.loc[:,'x2'],marker='x')
anamoly_points=plt.scatter(data.loc[:,'x1'][y_predict==-1],data.loc[:,'x2'][y_predict==-1],marker='o',facecolor='none',edgecolor='red',s=150)
plt.title('data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend((mormaly_points,anamoly_points),('mormaly_points','anamoly_points'))
plt.show()
```

```python
clf = EllipticEnvelope(contamination=0.02)
clf.fit(data)
y_predict = clf.predict(data)
fig2 = plt.figure(figsize=(10,5))
mormaly_points=plt.scatter(data.loc[:,'x1'],data.loc[:,'x2'],marker='x')
anamoly_points=plt.scatter(data.loc[:,'x1'][y_predict==-1],data.loc[:,'x2'][y_predict==-1],marker='o',facecolor='none',edgecolor='red',s=150)
plt.title('data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend((mormaly_points,anamoly_points),('mormaly_points','anamoly_points'))
plt.show()
```
